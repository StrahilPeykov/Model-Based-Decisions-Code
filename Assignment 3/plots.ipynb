```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("results/baseline_sweep_summary.csv")

# pivot for heatmaps
p = df.pivot(index="I0_grid", columns="X0", values="p_high").sort_index()
b = df.pivot(index="I0_grid", columns="X0", values="bistable").sort_index()

# p_high heatmap
plt.figure()
plt.imshow(p.values, origin="lower", aspect="auto")
plt.colorbar(label="p_high")
plt.xlabel("X0")
plt.ylabel("I0")
plt.xticks(range(len(p.columns)), [f"{x:.2f}" for x in p.columns], rotation=90)
plt.yticks(range(len(p.index)), [f"{i:.2f}" for i in p.index])
plt.title("Baseline: Probability of high adoption (p_high)")
plt.tight_layout()
plt.show()
plt.savefig("figures/Fig_P1_p_high_heatmap.png", dpi=200)
plt.close()

# bistable heatmap
plt.figure()
plt.imshow(b.values, origin="lower", aspect="auto", vmin=0, vmax=1)
plt.colorbar(label="bistable (0/1)")
plt.xlabel("X0")
plt.ylabel("I0")
plt.xticks(range(len(b.columns)), [f"{x:.2f}" for x in b.columns], rotation=90)
plt.yticks(range(len(b.index)), [f"{i:.2f}" for i in b.index])
plt.title("Baseline: Bistability region (two basins observed)")
plt.tight_layout()
plt.show()
plt.savefig("figures/Fig_P1_bistable_heatmap.png", dpi=200)
plt.close()

print("Saved:")
print(" - figures/Fig_P1_p_high_heatmap.png")
print(" - figures/Fig_P1_bistable_heatmap.png")
```
